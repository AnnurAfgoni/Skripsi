## Library

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split

from keras.models import Sequential, Model
from keras.layers import concatenate
from keras.layers import Dense, Input, Dropout

import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
```

## Shared Layer Model

```python
def model(n_input, n_target):
    """
    Fungsi untuk membuat model
    
    arguments:
        n_input = jumlah feature dari data input
        n_outut = jumlah target
        
    return:
        model
    """
    # initializer 
    initializer = tf.keras.initializers.GlorotUniform(seed=None)
    mse = tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error")
    aktivasi = "relu"

    # input layer
    visible = Input(shape=(n_input))
    
    # branch 1
    x1 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x1 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x1)
    x1 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x1)
    x1 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x1)
    
    # branch 2
    x2 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x2 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x2)
    x2 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x2)
    x2 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x2)
    
    # branch 3
    x3 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x3 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x3)
    x3 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x3)
    x3 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x3)
    
    # branch 4
    x4 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x4 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x4)
    x4 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x4)
    x4 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x4)
    
    # branch 5
    x5 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x5 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x5)
    x5 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x5)
    x5 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x5)
    
    # branch 6
    x6 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x6 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x6)
    x6 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x6)
    x6 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x6)
    
    # branch 7
    x7 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x7 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x7)
    x7 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x7)
    x7 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x7)
    
    # branch 8
    x8 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x8 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x8)
    x8 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x8)
    x8 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x8)
    
    # branch 9
    x9 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x9 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x9)
    x9 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x9)
    x9 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x9)
    
    # branch 10
    x10 = Dense(100, kernel_initializer = initializer, activation = aktivasi)(visible)
    x10 = Dense(75, kernel_initializer = initializer, activation = aktivasi)(x10)
    x10 = Dense(50, kernel_initializer = initializer, activation = aktivasi)(x10)
    x10 = Dense(n_target, kernel_initializer = initializer, activation = "tanh")(x10)


    # output
    output = concatenate([x1, x2, x3, x4, x5, x6, x7, x8, x9, x10])

    # model
    model = Model(inputs = visible, 
                  outputs = output)
    
    #compile model
    model.compile(loss = mse,
                  optimizer = 'nadam')
    
    return model
```

## Data

```python
# input data

feature = pd.read_csv("coef_fourier_kolom.csv")
feature.sample(5)
```

```python
target = pd.read_csv("coef_ekspansi.csv")
target.sample(5)
```

## Data Splitting

```python
instances_1, feature_dim = feature.shape
instances_2, target_dim = target.shape
```

```python
# split training and testing
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size = 0.2, random_state = 7)
```

## Scaling

```python
# TRAINING SCALING
scaler = StandardScaler()

# fit on the training dataset
scaler.fit(X_train)

# scale the training set
X_train = scaler.transform(X_train)

# scale the test dataset
X_test = scaler.transform(X_test)
```

```python
#TESTING SCALER
scaler = MinMaxScaler(feature_range = (-1, 1))

# fit on the test dataset
scaler.fit(y_train)

# scale the test set
y_train_dummy = scaler.transform(y_train)

# scale the test dataset
y_test_dummy = scaler.transform(y_test)
```

## Training

```python
# model 

model = model(feature_dim, 10)
```

```python
model.summary()
```

```python
# training 
history = model.fit(x = X_train, 
           y = y_train_dummy, 
           batch_size = 8, 
           shuffle = True, 
           validation_data = (X_test, y_test_dummy), 
           epochs = 250)
```

## Visualisasi

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
epochs = range(1, len(loss) + 1)
plt.figure(figsize = (13, 8))
plt.style.use("seaborn")

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.show()
```

## Prediction

```python
# prediksi
prediction = model.predict(X_test)
prediction = scaler.inverse_transform(prediction)
```

```python
prediksi = prediction.flatten()
real = y_test.to_numpy().flatten()
```

```python
df = {"predict": prediksi, "real": real}
pd.DataFrame(df)
```
