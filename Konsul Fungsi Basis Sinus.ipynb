```python
import numpy as np
import matplotlib.pyplot as plt
import math
import numpy.linalg as la
from matplotlib.pyplot import cm
```

## 2 Dimensi

```python
def psi(m, n):
    """
        LX = batas sumbu x
        LY = batas sumbu y
        m = indeks sumbu x
        n = indeks sumbu y
    """
    # fungsi gelombang
    psi = 2/np.sqrt(LX*LY) * np.sin(m*np.pi*x/LX) * np.sin(n*np.pi*y/LY)
    
    return psi
```

```python
def d2psi(m, n):
    
    psix = -0.5 * -(m*np.pi/LX)**2 * 2/np.sqrt(LX*LY) * np.sin(m*np.pi*x/LX) * np.sin(n*np.pi*y/LY)
        
    psiy = -0.5 * -(n*np.pi/LY)**2 * 2/np.sqrt(LX*LY) * np.sin(m*np.pi*x/LX) * np.sin(n*np.pi*y/LY)
    
    return psix, psiy
```

```python
# Jumlah grid
NX = 100
NY = 100

# panjang interval 
LX = 10
LY = 10

# Panjang grid
dx = LX/NX
dy = LY/NY

# titik tengah
x0 = LX/2
y0 = LY/2

# data input
X = np.linspace(0, LX, NX+1) 
Y = np.linspace(0, LY, NY+1)

x, y = np.meshgrid(X, Y)

# potensial OH
V = 0.5*(x - x0)**2 + 0.5*(y - y0)**2
```

```python
plt.contourf(x, y, V, cmap = cm.coolwarm)
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(aspect=8)
plt.gca().set_aspect('equal')
```

## Representasi Matriks

```python
rep_matriks = np.array([])

# baris dan kolom matriks
n_bar = 10
n_kol = 10

for m in range(1, n_bar+1):
    for n in range(1, n_kol+1):
        for p in range(1, n_bar+1):
            for q in range(1, n_kol+1):
        
                energi = 0.

                psi_kiri = psi(m, n)
                psi_kanan = psi(p, q)
                
                d2psix, d2psiy = d2psi(p, q)

                for i in range(1, NX):
                    for j in range(1, NY):
                        
                        energi += psi_kiri[i,j]*(d2psix[i,j] + d2psiy[i,j]) + psi_kiri[i,j]*V[i,j]*psi_kanan[i,j]

                energi *= dx*dy
                print(f"energi_{m}{n}_{p}{q} = {energi}")
                rep_matriks = np.append(rep_matriks, energi)
```

```python
rep_matriks2 = rep_matriks.reshape(n_bar*n_kol, n_bar*n_kol)
rep_matriks2
```

```python
# menghitung nilai eigen
w, v = la.eig(rep_matriks2)
```

```python
print(w)
```

```python
w1 = np.sort(w)
w1
```

```python
# energi
gs = np.min(w)
igs = np.argmin(w)

print("energi = ", gs)
print("indeks = ", igs)
```

```python
# koefisien ekspansi ground state
cn = v[:, igs]
cn1 = cn.reshape(n_bar, n_kol)
print(cn1)
```

```python
gelombang = 0.

for i in range(n_bar):
    for j in range(n_kol):
        gelombang += cn1[i,j]*psi(i+1, j+1)
```

```python
plt.title("Fungsi Gelombang Dasar Osilator Harmonik 2D")
plt.contourf(x, y, gelombang**2, cmap = cm.coolwarm)
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(aspect=8)
plt.gca().set_aspect('equal')

print("energi = ", gs)
```
